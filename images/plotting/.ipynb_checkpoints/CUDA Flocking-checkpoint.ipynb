```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
# with visualization
naiveBoids = np.linspace(10000, 100000, 8)
naiveFPS = [180, 92, 64, 46, 27, 20, 15, 12]
# without visualization
naiveBoidsV = np.linspace(10000, 100000, 8)
naiveFPSV = [207, 102, 68, 49, 29, 20, 15, 12]

naiveFig, naiveAxes = plt.subplots()

naiveAxes.plot(naiveBoids, naiveFPS, label="With Visualization", marker='o', markerfacecolor="yellow", markeredgecolor="green")
naiveAxes.plot(naiveBoidsV, naiveFPSV, label="Without Visualization", marker='o', markerfacecolor="yellow", markeredgecolor="green")
naiveAxes.get_xaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
naiveAxes.yaxis.set_ticks(np.arange(0, 220, 10))
naiveAxes.xaxis.set_ticks(np.arange(10000, 110000, 15000))
naiveAxes.set_xlabel('Number of Boids') # Notice the use of set_ to begin methods
naiveAxes.set_ylabel('FPS')
naiveAxes.set_title('Naive Approach')
naiveAxes.axhline(y=30, color='r', linestyle='--',alpha=0.5)
naiveAxes.axhline(y=60, color='g', linestyle='--',alpha=0.5)
naiveAxes.legend()

# with visualization
uniformBoids = np.linspace(100000, 700000, 7)
uniformFPS = [440, 200, 137, 60, 45, 27, 17]
# without visualization
uniformBoidsV = np.linspace(100000, 700000, 7)
uniformFPSV = [600, 300, 145, 80, 48, 28, 17]

uniformFig, uniformAxes = plt.subplots()

uniformAxes.plot(uniformBoids, uniformFPS, label="With Visualization", marker='o', markerfacecolor="yellow", markeredgecolor="green")
uniformAxes.plot(uniformBoidsV, uniformFPSV, label="Without Visualization", marker='o', markerfacecolor="yellow", markeredgecolor="green")
uniformAxes.get_xaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
uniformAxes.yaxis.set_ticks(np.arange(0, 650, 50))
#uniformAxes.xaxis.set_ticks(np.arange(10000, 110000, 15000))
uniformAxes.set_xlabel('Number of Boids') # Notice the use of set_ to begin methods
uniformAxes.set_ylabel('FPS')
uniformAxes.set_title('uniform Approach')
uniformAxes.axhline(y=30, color='r', linestyle='--',alpha=0.5)
uniformAxes.axhline(y=60, color='g', linestyle='--',alpha=0.5)
uniformAxes.legend()


# with visualization
coherentBoids = np.linspace(1000000, 2500000, 4)
coherentFPS = [95, 50, 30, 19]
# without visualization
coherentBoidsV = np.linspace(1000000, 2500000, 4)
coherentFPSV = [108, 53, 31, 20]

coherentFig, coherentAxes = plt.subplots()

coherentAxes.plot(coherentBoids, coherentFPS, label="With Visualization", marker='o', markerfacecolor="yellow", markeredgecolor="green")
coherentAxes.plot(coherentBoidsV, coherentFPSV, label="Without Visualization", marker='o', markerfacecolor="yellow", markeredgecolor="green")
coherentAxes.yaxis.set_ticks(np.arange(0, 110, 10))
coherentAxes.set_xlabel('Number of Boids') # Notice the use of set_ to begin methods
coherentAxes.set_ylabel('FPS')
coherentAxes.set_title('coherent Approach')
coherentAxes.axhline(y=30, color='r', linestyle='--',alpha=0.5)
coherentAxes.axhline(y=60, color='g', linestyle='--',alpha=0.5)
coherentAxes.legend()

# with visualization
blockSize = [128, 256, 512, 1024]
blockFPS = [850, 859, 860, 900]

blockFig, blockAxes = plt.subplots()

blockAxes.plot(blockSize, blockFPS, marker='o', markerfacecolor="yellow", markeredgecolor="green")
blockAxes.xaxis.set_ticks(np.arange(0, 1088, 128))
blockAxes.set_xlabel('Block Size') 
blockAxes.set_ylabel('FPS')
blockAxes.set_title('Block Size vs FPS')



naiveFig.savefig("naive.png")
uniformFig.savefig("uniform.png")
coherentFig.savefig("coherent.png")
blockFig.savefig("blocksize.png")
```
