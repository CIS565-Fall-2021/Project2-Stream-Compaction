```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

arraySize = [1024, 4096, 16384, 65536]
CPUScan = [0.0014,  0.0054, 0.0213,  0.0917]
GPUEfficient = [0.017408,  0.017408, 0.018432, 0.099712]
GPUNaive = [0.016384, 0.016384,  0.018432, 0.075776]
Thrust = [0.059072,  0.068608,  0.05632, 0.08601]

scanFig, scanAxes = plt.subplots()

scanAxes.plot(arraySize, CPUScan, label="CPU Scan", marker='o')
scanAxes.plot(arraySize, GPUEfficient, label="GPU Efficient", marker='o')
scanAxes.plot(arraySize, GPUNaive, label="GPU Naive", marker='o')
scanAxes.plot(arraySize, Thrust, label="Thrust", marker='o')
scanAxes.set_xlabel('Array Size') # Notice the use of set_ to begin methods
scanAxes.set_ylabel('Time (ms)')
scanAxes.set_title('Array Size vs Time')
scanAxes.legend()


scanFig.savefig("../scan.png")
```
